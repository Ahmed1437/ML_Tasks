## 3. try to make a plot like given one
- you need to choose the top 5 car brands only from data
- use `make` in x axis and `ghgScore` in y axis 

2.jpg

```python
list = df['make'].value_counts().index[:5]
```

```python
new_df = df[df['make'].isin(list)]
```

```python
plt.figure(figsize = [15,10])
sns.barplot(data = new_df, x = 'make', y = 'ghgScore', hue = 'VClass', ci = False)
plt.show()
```
